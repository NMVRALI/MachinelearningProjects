```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler

# Generating a synthetic dataset
# the number of samples should be 300 with noice 0.05
X, y = make_moons(n_samples=300, noise=0.05, random_state=42)
# Normalize X
SS = StandardScaler()
X = SS.fit_transform(X)


# Plotting the dataset
plt.scatter(X[:,0],X[:,1] , s=10,color='blue')
plt.title("Synthetic Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

## 1. Dataset

**Dataset Overview:**

In this section, we generate a synthetic dataset using the `make_moons` function from `sklearn.datasets`. This dataset consists of 300 samples with some noise (0.05) and presents a non-linear structure. It will be used to test the performance of various clustering algorithms.

We then normalize the dataset using `StandardScaler` to ensure that the features are on the same scale, which can help improve the performance of some clustering algorithms.


## 2. K-means Clustering

**K-means Overview:**

In this section, we apply the K-means clustering algorithm to the synthetic dataset. K-means is an iterative algorithm that partitions the dataset into `k` clusters by minimizing the sum of squared distances between data points and their assigned cluster centers.

We set `n_clusters=2` since we know the data has two moon-like structures. The resulting plot will show how well K-means separates the two clusters.


```python
from sklearn.cluster import KMeans

# K-means clustering
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# Plotting K-means clustering results
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels,cmap='viridis', s=50)
plt.title("K-means Clustering")
plt.show()
```

## 3. Agglomerative Clustering

**Agglomerative Clustering Overview:**

Here, we use the Agglomerative Clustering algorithm, which is a hierarchical clustering method. It works by merging the closest clusters step by step until the desired number of clusters is reached. We set `n_clusters=2` to match the two moon-like structures in the dataset.

Agglomerative clustering doesn’t require a predefined number of clusters in some variants, but in this case, we fix it to 2 for comparison with K-means. The plot will help us evaluate its performance.


```python
from sklearn.cluster import AgglomerativeClustering

# Agglomerative clustering
agglo = AgglomerativeClustering(n_clusters=2)
agglo_labels = agglo.fit_predict(X)

# Plotting Agglomerative clustering results
plt.scatter(X[:, 0], X[:, 1], c=agglo_labels,cmap='viridis', s=50)
plt.title("Agglomerative Clustering")
plt.show()
```

## 4. DBSCAN

**DBSCAN Overview:**

In this section, we apply DBSCAN (Density-Based Spatial Clustering of Applications with Noise), a clustering algorithm that groups points based on their density. It is particularly good for identifying clusters of varying shapes and sizes, as well as noise (outliers). 

We set `eps=0.3` and `min_samples=5` to control the density criteria. DBSCAN can also detect noise points, which will be labeled as `-1` in the resulting plot.


```python
from sklearn.cluster import DBSCAN

# DBSCAN clustering
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

# Plotting DBSCAN clustering results
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis', s=50)
plt.title("DBSCAN Clustering")
plt.show()
```

## Conclusion and Final Thoughts

In this assignment, we explored and compared three popular clustering algorithms: K-means, Agglomerative Clustering, and DBSCAN. Here's a brief summary of the findings:

- **K-means** performed well in situations where the data is relatively well-separated and spherical. However, it may struggle with non-convex clusters or noise.
- **Agglomerative Clustering** offers a more flexible approach compared to K-means, as it can handle hierarchies and doesn’t require a predefined number of clusters. It performed well in identifying clusters that were not as well-separated.
- **DBSCAN** was able to identify non-linearly shaped clusters and also detect noise points. It excels when dealing with data that contains outliers and clusters of varying shapes.

Each algorithm has its strengths and is suitable for different types of datasets. K-means is efficient and works well for datasets with spherical clusters, Agglomerative Clustering is ideal for hierarchical data, and DBSCAN is particularly useful for identifying noise and non-linear structures.
